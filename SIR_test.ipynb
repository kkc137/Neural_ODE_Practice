```julia
using DifferentialEquations,ParameterizedFunctions, Plots
```

```julia

function sir_ode1(du,u,p,t)
    S,I,R = u
    b,k = p
    du[1] = -b*S*I
    du[2] = b*S*I-k*I
    du[3] = k*I
end
```

```julia
parms = [0.1,0.05]
init = [0.99,0.01,0.0]
tspan = (0.0,200.0)
sir_prob = ODEProblem(sir_ode1,init,tspan,parms)
```

```julia
sir_sol = solve(sir_prob,saveat = 0.1) 
```

```julia
plot(sir_sol,xlabel="Time",ylabel="Number") # for visualization
```
